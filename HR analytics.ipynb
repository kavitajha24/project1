```python
#making the skew less than or equal to 0.55 for better prediction and plotting Normal distribution
skew=('MonthlyIncome','NumCompaniesWorked','YearsSinceLastPromotion','TotalWorkingYears','JobLevel')
for col in skew :
    if data.skew().loc[col]>0.55:
        data[col] = np.log1p(data[col])
```

```python
#train and testing phase
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.35,random_state=123)
```

```python
#shape of x_train,y_train
print('Shape of x_train is:', x_train.shape)
print('Shape of y_train is:', y_train.shape)
print('Shape of x_test is:', x_test.shape)
print('Shape of y_test is:', y_test.shape)
```

###### model1 
Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score,roc_curve
from sklearn.model_selection import GridSearchCV
```

```python
lr=LogisticRegression()  #instantiate
lr.fit(x_train,y_train)
```

```python
y_pred_lr=lr.predict(x_test)
```

```python
train1=lr.score(x_train,y_train)
print('Accuracy of training data of logistic regression is :',train1)
```

```python

accuracy_score(y_test,y_pred_lr)
```

accraccy of testing data of logistic regression is 90 %

```python
print(classification_report(y_test,y_pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lr=LogisticRegression()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
kn=KNeighborsClassifier()
```

```python

clf=RandomForestClassifier()
params={'n_estimators':[13,15],
       'criterion':['entropy'],
       'max_depth':[10,15],
       'min_samples_split':[10,11],
       'min_samples_leaf':[5,6]
       }

grd=GridSearchCV(clf,param_grid=params)
grd.fit(x_train,y_train)
print('best_params=>',grd.best_params_)

clf=grd.best_estimator_

clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)

rf_conf_mat=confusion_matrix(y_test,y_pred)

print('\ncon_mat','\n',rf_conf_mat)
print('\n Accuracy rf normal',accuracy_score(y_test,y_pred))
```

##### Lets plot ROC AUC Curve to choose best model


```python
lr=LogisticRegression()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
kn=KNeighborsClassifier()
```

```python
#training with all classifier
lr.fit(x_train,y_train)
dt.fit(x_train,y_train)
rf.fit(x_train,y_train)
kn.fit(x_train,y_train)

print('All Models are trained')
```

```python
#all models scores are captured
lr.score(x_test,y_test)
dt.score(x_test,y_test)
rf.score(x_test,y_test)
kn.score(x_test,y_test)

print('All Models test score captured')
```

##### ets check ROC AUC curve for the fitted model

```python
from sklearn.metrics import roc_auc_score,roc_curve
from sklearn.metrics import plot_roc_curve
disp=plot_roc_curve(dt,x_test,y_test)
plot_roc_curve(lr,x_test,y_test,ax=disp.ax_)
plot_roc_curve(kn,x_test,y_test,ax=disp.ax_)
plot_roc_curve(rf,x_test,y_test,ax=disp.ax_)
plt.legend(prop={'size':11},loc='lower right')
plt.show()      
```
