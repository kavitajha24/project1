```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
url='https://raw.githubusercontent.com/dsrscientist/DSData/master/happiness_score_dataset.csv'
data=pd.read_csv(url)
data
```

```python
data=data.drop(columns=['Country'])
data
```

```python
data=data.drop(columns=['Region'])
data
```

```python
data=data.drop(columns=['Happiness Rank'])
data
```

```python
plt.figure(figsize=(20,20))
plotnumber=1
for column in data:
    if plotnumber<=9:
        ax=plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```

```python
data.skew()
```

```python
#detecting outliers and deal withthem 
from scipy.stats import zscore
```

```python
z=np.abs(zscore(data))
z.shape
```

```python
filtering_entry=((z<3).all(axis=1))
data=data[filtering_entry]
data.shape
```

```python
data.head()
```

```python
plt.figure(figsize=(20,20))
plotnumber=1
for column in data:
    if plotnumber<=9:
        ax=plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column, fontsize=20)
    plotnumber+=1
plt.show()
```

```python
x=data.drop(columns=['Happiness Score'])
y=data['Happiness Score']
```

```python
plt.figure(figsize=(20,20))
plotnumber=1
for column in x:
    if plotnumber<=15:
        ax=plt.subplot(5,3,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column)
        plt.ylabel('Happiness Score')
    plotnumber+=1
plt.show()
```

```python
#check multicollinearity using heatmap
data_corr=data.corr().abs()
plt.figure(figsize=(22,15))
sns.heatmap(data_corr,annot=True,annot_kws={'size':15})
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled.shape[1]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif['features']=x.columns
vif
```

none of them is not equal to more than 5 that means multicollinearity not exist

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=123)
y_train.head()
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
data.tail()
```

```python
print("Happiness Score",lr.predict(scaler.transform([[0.05141,0.46534,0.77115,0.15185,0.46866,0.17922,0.20165,1.141723]])))
```

```python
print('Happiness Score',lr.predict(scaler.transform([[0.04324,0.25812,0.85188,0.27125,0.39493,0.12832,0.21747,1.46494]])))
```

```python
#adjusted R2
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_test,y_pred)
plt.xlabel("actual Happiness Score")
plt.ylabel("Predicted Happiness score")
plt.title("Actual vs Predicted")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
