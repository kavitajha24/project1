###### Splitting the data

```python
x=data.drop(columns=['RainTomorrow'])
y=data['RainTomorrow']
```

```python
#check shape of x and y 
print('Shape of x is :',x.shape,'\n\n') 
print('Shape of y is :',y.shape,'\n\n')
```

```python
#check relationship between x and y 
plt.figure(figsize=(20,20))
plotnumber=1
for column in x:
    if plotnumber<=16:
        ax=plt.subplot(7,4,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=10)
        plt.ylabel('RainTomorrow')
    plotnumber+=1
plt.show()
```

```python
#importing standard scler to scale the x 
from sklearn.preprocessing import StandardScaler
```

```python
s1=StandardScaler()  #instantiate
scaled1=s1.fit_transform(x)
scaled1.shape[1]
```

```python
#import train_test_split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled1,y,test_size=0.40,random_state=123)
```

###### 1st Model Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score,roc_curve
```

```python
lr=LogisticRegression()  #instantiate
lr.fit(x_train,y_train)
```

```python
#Trainig Accuracy of Logistic Regression Model
lr.score(x_train,y_train)
```

```python
y_pred_lr=lr.predict(x_test)
```

```python
#testing accuracy of Logistic Regression Model
accuracy_score(y_test,y_pred_lr)
```

Training score of Logictis Regression Model- 81%

Testing score of Logictis Regression Model- 80%

```python
#confusion matric of logistic Regression Model
confusion_matrix(y_test,y_pred_lr)
```

```python
#classification report
print(classification_report(y_test,y_pred_lr))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#cross val score for Logostoc Regression Model
cross_val_score(lr,scaled1,y,cv=5).mean()
```

Cross validation score for Logistic Regression Model is 79% that means model is not over fit

```python
#plt_roc_curve for Logistic Regression Model
fpr,tpr,thresholds=roc_curve(y_test,y_pred_lr)
```

```python
print('False Positive Rates are :',fpr)
print('True Positive Rates are :',tpr)
print('Thresholds are :',thresholds)
```

```python
plt.plot(fpr,tpr,color='orange',label='ROC')
plt.plot([0,1],[0,1],color='darkblue')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Logistic Regression model')
plt.legend()
plt.show()
```

This much area covers in Logistic Regression Model

```python
#conclusions for Logistic Regression Model
a=np.array(y_test)
predicted=np.array(lr.predict(x_test))
con=pd.DataFrame({'Original':a,'Predicted':predicted},index=range(len(a)))
con
```
